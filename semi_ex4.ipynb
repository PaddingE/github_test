```python
import pandas as pd
import numpy as np
```

```python
df_data1 = pd.read_excel('./전국_연령별_장애유형별_성별_등록장애인수_20230515034425.xlsx')
df_data2 = pd.read_excel('./전국_연령별_장애유형별_성별_등록장애인수_20230515034348.xlsx')
df_data3 = pd.read_excel('./전국_연령별_장애유형별_성별_등록장애인수_20230515033734.xlsx')
```

```python
print(df_data1.shape)

df_data1

df_data1_short_2017 = df_data1.loc[:26,'2017']
df_data1_short_2018 = df_data1.loc[:26,'2018']

df_data1_short_2017
df_data1_short_2018

df_data1_short = pd.concat([df_data1_short_2017,df_data1_short_2018], axis= 1)

print(df_data1_short)
```

```python
print(df_data2.shape)

df_data2_short = df_data2.loc[:26,['2019','2020']]

df_data2_short
```

```python
df_data3_short = df_data3.loc[:26,['2021','2022']]

df_data3_short
```

```python
df_data_short = pd.concat([df_data1_short,df_data2_short,df_data3_short], axis= 1)

df_data_short
```

```python
df_data_short.iloc[3]
```

```python
df_total = df_data_short.iloc[3] + df_data_short.iloc[9] + df_data_short.iloc[15] + df_data_short.iloc[21]

df1 = df_data_short.iloc[3] / df_total * 100
df2 = df_data_short.iloc[9] / df_total * 100
df3 = df_data_short.iloc[15] / df_total * 100
df4 = df_data_short.iloc[21] / df_total * 100

df_data_final = pd.concat([df1 , df2 , df3,df4],axis= 1)
df = df_data_final.T
df
```

```python
import matplotlib.pyplot as plt
```

```python

plt.figure(figsize=(10, 5))
plt.plot(df1, marker='o',label = ['1-4'])
plt.plot(df2, marker='o',label = ['5-9'])
plt.plot(df3, marker='o',label = ['10-14'])
plt.plot(df4, marker='o',label = ['15-19'])
plt.xlabel('age')
plt.ylabel('num')
plt.legend()
plt.show()
```
