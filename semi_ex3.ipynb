```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_data_main = pd.read_excel('./시도_초혼연령별_혼인_20230515053319.xlsx')

df_data_main
```

```python
df_data_main['시도별'].fillna(method= 'ffill',inplace= True)

df_data_main
```

```python
df_data_copy = df_data_main.copy()

df_data_all = df_data_copy.iloc[0:16].copy()

df_data_all.drop('시도별',axis = 1,inplace=True)
df_data_all.drop([2,3],inplace=True)
df_data_all.reset_index(inplace=True)
df_data_all.drop('index',axis = 1,inplace=True)

for i in range(8,14):
    df_data_all.drop([i],inplace=True)
    
    
print(df_data_all)

#df_data_all.to_excel("./수정용.xlsx")
```

```python
df_for_cal = df_data_all.copy()
df_for_cal.drop('연령별', axis= 1, inplace=True)
print(df_for_cal.shape)
df_data_20s_male = (df_for_cal.iloc[2, 0::2] + df_for_cal.iloc[3, 0::2]) / df_for_cal.iloc[1, 0::2] *100
df_data_30s_male = (df_for_cal.iloc[4, 0::2] + df_for_cal.iloc[5, 0::2]) / df_for_cal.iloc[1, 0::2] *100
df_data_40s_male = (df_for_cal.iloc[6, 0::2] + df_for_cal.iloc[7, 0::2]) / df_for_cal.iloc[1, 0::2] *100

dict_data = {'20대': df_data_20s_male, '30대': df_data_30s_male, '40대': df_data_40s_male}

df = pd.DataFrame(dict_data)

df
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df, marker='o',label= ['20s_male', '30s_male','40s_male'])

plt.xlabel('year')
plt.ylabel('ratio')
plt.legend()
plt.show()
```
